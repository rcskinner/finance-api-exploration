# Yahoo Finance / yfinance Python Package

Exploring the options available with yahoo finance and the yfinance python package

[yfinance](https://github.com/ranaroussi/yfinance)

Initial Impressions: 
- Plug and play with a given ticker is nice
- Returns as pandas dataframe with no need for conversions

Cons: 
- dividends method seems inconsistent if ticker.history() has been run. The dividends shown will then only be shown for dates within the specified time within the history
- Package testing seems to be lacking
- Limited development lately (~300 open issues and no resolution or work)
- No earnings data available? Bug or change in the Yahoo Finance API?

Major Methods: 
- history - Returns the open/high/low/close/volume for the specified asset over a time horizon
- actions - Shows any dividends or stock splits that occured
